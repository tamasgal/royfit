# DU-1 ROyFit Reconstruction (work in progress)

ROyFit uses the BBfit approach to reconstruct the zenith angle of muon tracks. This is a quick summary of its current performance on the DU-1, measured on ~20k atmospheric muon events (from Rosa).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import pickle
from pprint import pprint
```

```python
plt.xkcd()
from matplotlib import rcParams
rcParams['axes.grid'] = False
rcParams['xtick.major.size'] = 0
rcParams['ytick.major.size'] = 0
```

The royfit_stats.pickle is generated by ROyFit and is pickle dump of a Python dictionary. It holds several reconstruction parameters, which can be accessed using the corresponding keys. The values are stored in a flat list keeping the original ordering.

```python
with open('royfit_stats_firstt3hits_tot30.pickle') as royfit_stats:
    stats = pickle.load(royfit_stats)
```

### Fit parameters:

```python
pprint(stats['fit_parameters'])
```

### Fit results:

```python
q = stats['quality_parameter']
ang_err = stats['angular_error']
mc_zenith = stats['mc_zenith']
reco_zenith = stats['reco_zenith']
number_of_events = stats['number_of_events']
number_of_vaild_fits = stats['number_of_vaild_fits']
dc = stats['dc']
uz = stats['uz']
zc = stats['zc']
tc = stats['tc']
dc_err = stats['dc_err']
uz_err = stats['uz_err']
zc_err = stats['zc_err']
tc_err = stats['tc_err']
```

## Reconstruction Quality

The overall reconstruction quality is OK, however there is a tendency to "pull" the reconstructed track to the horizon, shown in the histogram below. This is probably due to the weighting term of the quality function (see the histogram at the end, showing the reconstructed distance to line).

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

plt.suptitle("ROyFit on {0} MUPAGE events with {1} valid fits" \
             .format(number_of_events, number_of_vaild_fits), size=16, y=1.01)

hist2d = axes[0].hist2d(mc_zenith, reco_zenith, bins=60)
axes[0].set_xlabel('MC zenith [degree]')
axes[0].set_ylabel('reco zenith [degree]')
xx = np.linspace(min(mc_zenith), max(mc_zenith), 10)
axes[0].plot(xx, xx, 'r--')

fig.colorbar(hist2d[3], ax=axes[0])

#extent = [yedges[0], yedges[-1], xedges[0], xedges[-1]]
#im = axes[0].imshow(H, cmap=plt.cm.jet, extent=extent)
#fig.colorbar(im, ax=axes[0])

axes[1].hist(ang_err, bins=100, normed=True)
axes[1].set_xlabel('(MC zenith - reco zenith) [deg]')
axes[1].set_ylabel("")




mean = np.mean(ang_err)
variance = np.var(ang_err)
sigma = np.sqrt(variance)
xx = np.linspace(min(ang_err), max(ang_err),100)
#axes[1].plot(xx, mlab.normpdf(xx, mean, sigma))
```

## Quality Parameter

The quality parameter is defined by $q=Q/n_{freedoms}$,where Q is the value of the minimised function) and n the number of freedoms (n=4 in our case).

Below is a histogram of q vs angular_error for reconstructions, where q is less then max_q. Unfortunatelly there is no suitable cut yet.

```python
max_q = 4
q_ang_err_filtered = [(_q, _ang_err) for _q, _ang_err in zip(q, ang_err) if _q < max_q]
q_filtered, ang_err_filtered = zip(*q_ang_err_filtered)

plt.hist2d(q_filtered, ang_err_filtered, bins=50)
plt.title("Q vs angular_error\nfrom {0} MUPAGE events with {1} valid fits"
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel('Q/4')
plt.ylabel("(MC zenith - reco zenith) [degree]")
plt.colorbar()
```

## Calculated Distances to Line

There is an excessive pull to the limits of dc (2, 100) towards the detector line. The original BBfit implementation uses a weighting term, which tries to balance the correlation between the PMT charges and the expected photons. In case of the KM3NeT DOMs, there is no charge, so ROyFit uses the number of PMTs hit within 15ns after the first T3 hit on the DOM. The weighting term is however not perfect yet and still under investigation.

```python
bins = 100

plt.hist(dc, bins=bins)
plt.xlabel('dc [m]')
plt.ylabel('count')
plt.title('ROyFit Calculated Distances to Line')
```

```python
plt.hist2d(dc, ang_err, bins=50)
plt.title("dc vs angular_error\nfrom {0} MUPAGE events with {1} valid fits"
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel('dc [m]')
plt.ylabel("(MC zenith - reco zenith) [degree]")
plt.colorbar()
```
