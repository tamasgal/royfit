# DU-1 reco analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

The royfit_stats.pickle is generated by ROyFit and is pickle dump of a Python dictionary. It holds several reconstruction parameters, which can be accessed using the corresponding keys. The values are stored in a flat list keeping the original ordering.

```python
with open('../royfit_stats.pickle') as royfit_stats:
    stats = pickle.load(royfit_stats)
```

```python
q = stats['quality_parameter']
ang_err = stats['angular_error']
mc_zenith = stats['mc_zenith']
reco_zenith = stats['reco_zenith']
number_of_events = stats['number_of_events']
number_of_vaild_fits = stats['number_of_vaild_fits']
dc = stats['dc']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

plt.suptitle("ROyFit on {0} MUPAGE events with {1} valid fits" \
             .format(number_of_events, number_of_vaild_fits), size=16, y=1.01)

axes[0].hist2d(mc_zenith, reco_zenith, bins=30)
axes[0].set_xlabel('MC zenith [degree]')
axes[0].set_ylabel('reco zenith [degree]')
xx = np.linspace(min(mc_zenith), max(mc_zenith), 10)
axes[0].plot(xx, xx, 'r--')

axes[1].hist(ang_err, bins=40)
axes[1].set_xlabel('(MC zenith - reco zenith) [deg]')
axes[1].set_ylabel("count")
```

```python
plt.hist([x for x in q if x < 100], bins=40)
plt.title("ROyFit on {0} MUPAGE events with {1} valid fits - Q/4" \
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel("Q/4")
plt.ylabel("count")
```

Pick only those angular_error and quality_parameter tuples, where quality_parameter is less then max_q. Otherwise the histogram is crappy.

```python
max_q = 20
q_ang_err_filtered = [(_q, _ang_err) for _q, _ang_err in zip(q, ang_err) if _q < max_q]
q_filtered, ang_err_filtered = zip(*q_ang_err_filtered)
```

```python
plt.hist2d(q_filtered, ang_err_filtered, bins=50)
plt.title("Q vs angular_error\nfrom {0} MUPAGE events with {1} valid fits"
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel('Q/4')
plt.ylabel("(MC zenith - reco zenith) [degree]")
plt.colorbar()
```

```python
plt.hist(dc, bins=40)
plt.title("calculated distance from line\nfrom {0} MUPAGE events with {1} valid fits"
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel('dc [m]')
plt.ylabel("count")
```
