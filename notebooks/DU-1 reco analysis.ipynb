# DU-1 reco analysis

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

The royfit_stats.pickle is generated by ROyFit and is pickle dump of a Python dictionary. It holds several reconstruction parameters, which can be accessed using the corresponding keys. The values are stored in a flat list keeping the original ordering.

```python
with open('../royfit_stats.pickle') as royfit_stats:
    stats = pickle.load(royfit_stats)
```

```python
q = stats['quality_parameter']
ang_err = stats['angular_error']
mc_zenith = stats['mc_zenith']
reco_zenith = stats['reco_zenith']
number_of_events = stats['number_of_events']
number_of_vaild_fits = stats['number_of_vaild_fits']
dc = stats['dc']
uz = stats['uz']
zc = stats['zc']
tc = stats['tc']
dc_err = stats['dc_err']
uz_err = stats['uz_err']
zc_err = stats['zc_err']
tc_err = stats['tc_err']
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

plt.suptitle("ROyFit on {0} MUPAGE events with {1} valid fits" \
             .format(number_of_events, number_of_vaild_fits), size=16, y=1.01)

axes[0].hist2d(mc_zenith, reco_zenith, bins=60)
axes[0].set_xlabel('MC zenith [degree]')
axes[0].set_ylabel('reco zenith [degree]')
xx = np.linspace(min(mc_zenith), max(mc_zenith), 10)
axes[0].plot(xx, xx, 'r--')

axes[1].hist(ang_err, bins=100)
axes[1].set_xlabel('(MC zenith - reco zenith) [deg]')
axes[1].set_ylabel("count")
```

```python
plt.hist([x for x in q if x < 100], bins=100)
plt.title("ROyFit on {0} MUPAGE events with {1} valid fits - Q/4" \
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel("Q/4")
plt.ylabel("count")
```

# Quality Parameter

The quality parameter is defined by

$$q=Q/n_{freedoms}$$

where Q is the value of the minimised function) and n the number of freedoms (n=4 in our case).

Now we pick only those angular_error and q tuples, where q is less then max_q, to limit the x-range.

```python
max_q = 20
q_ang_err_filtered = [(_q, _ang_err) for _q, _ang_err in zip(q, ang_err) if _q < max_q]
q_filtered, ang_err_filtered = zip(*q_ang_err_filtered)
```

```python
plt.hist2d(q_filtered, ang_err_filtered, bins=50)
plt.title("Q vs angular_error\nfrom {0} MUPAGE events with {1} valid fits"
          .format(number_of_events, number_of_vaild_fits),
          y=1.04)
plt.xlabel('Q/4')
plt.ylabel("(MC zenith - reco zenith) [degree]")
plt.colorbar()
```

# Fitting errors

These are the errors for the fitted parameters, calculated by iminuit.

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 8))

plt.suptitle("ROyFit minimiser errors", size=16, y=1.01)

bins = 100

axes[0][0].hist(dc_err, bins=bins)
axes[0][0].set_xlabel('dc_err [m]')
axes[0][0].set_ylabel('count')

axes[0][1].hist(uz_err, bins=bins)
axes[0][1].set_xlabel('uz_err [rad]')
axes[0][1].set_ylabel("count")

axes[1][0].hist(zc_err, bins=bins)
axes[1][0].set_xlabel('zc_err [m]')
axes[1][0].set_ylabel('count')

axes[1][1].hist(tc_err, bins=bins)
axes[1][1].set_xlabel('tc_err [ns]')
axes[1][1].set_ylabel("count")
```
